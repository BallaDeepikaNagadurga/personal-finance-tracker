```python
from abc import ABC, abstractmethod

class JuiceMachine(ABC):
    @abstractmethod
    def make_juice(self):
        pass

class OrangeJuiceMachine(JuiceMachine):
    def make_juice(self):
        print("Making orange juice...")

class AppleJuiceMachine(JuiceMachine):
    def make_juice(self):
        print("Making apple juice...")

orange = OrangeJuiceMachine()
apple = AppleJuiceMachine()

orange.make_juice()
apple.make_juice()
```
